```python
data = open("test.csv")

x = []
y = []

for line in data:
    
    line = line.strip()
    line = line.split(",")
    x.append(line[0])    
    y.append(line[1])       

wavelength = [float(i) for i in x[1:]]
flux = [float(i) for i in y[1:]]

```

```python
import matplotlib.pyplot as plt

plt.scatter(wavelength, flux)

index = flux.index(max(flux))
lambda_max = wavelength[index]

plt.scatter(lambda_max, max(flux))

lambda_max
```
